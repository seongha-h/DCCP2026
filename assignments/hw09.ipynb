```python
import csv
import json
import logging


logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)s:%(message)s"
)


def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    else:
        return "F"


def make_report(csv_path: str, json_path: str):
    report = []

    try:
        with open(csv_path, "r", encoding="utf-8") as f:
            reader = csv.DictReader(f)

            for row in reader:
                name = row["이름"]
                student_id = row["학번"]

                try:
                    korean = int(row["국어"]) if row["국어"] != "" else None
                    math = int(row["수학"]) if row["수학"] != "" else None
                    english = int(row["영어"]) if row["영어"] != "" else None

                    if korean is None or math is None or english is None:
                        avg = None
                        grade = None
                    else:
                        avg = round((korean + math + english) / 3, 1)
                        grade = get_grade(avg)

                    student = {
                        "이름": name,
                        "학번": student_id,
                        "점수": {
                            "국어": korean,
                            "수학": math,
                            "영어": english
                        },
                        "평균": avg,
                        "등급": grade
                    }

                    report.append(student)

                except ValueError:
                    logging.warning(f"{name} 학생의 점수 데이터가 올바르지 않습니다.")

        with open(json_path, "w", encoding="utf-8") as f:
            json.dump(report, f, ensure_ascii=False, indent=2)

        logging.info("학생 성적 리포트가 성공적으로 생성되었습니다.")

    except FileNotFoundError:
        logging.warning("CSV 파일을 찾을 수 없습니다.")

    except UnicodeDecodeError:
        logging.error("CSV 파일의 인코딩이 올바르지 않습니다.")
```

```python
make_report("students.csv", "report.json")
```

```python
class InvalidJamoError(ValueError):
    pass
def classify_jamo(c: str) -> str:
    if not isinstance(c, str):
        raise TypeError("입력은 문자열이어야 합니다.")

    if len(c) != 1:
        raise ValueError("입력은 한 글자여야 합니다.")

    code = ord(c)

    if 0x3131 <= code <= 0x314E:
        return "자음"
    elif 0x314F <= code <= 0x3163:
        return "모음"
    else:
        raise InvalidJamoError("한글 자모가 아닙니다.")


inputs = ["ㄱ", "ㅏ", "ㄲ", "가", "AB", 5, "ㅎ", "ㅣ", ""]

for x in inputs:
    try:
        result = classify_jamo(x)
        print(f"{x}: {result}")

    except TypeError as e:
        print(f"[TypeError] {e}")

    except InvalidJamoError as e:
        print(f"[InvalidJamoError] {e}")

    except ValueError as e:
        print(f"[ValueError] {e}")
```
